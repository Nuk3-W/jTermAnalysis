```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Assuming your CSV file is named 'nuclear_data.csv' and has columns 'year', 'country', and 'Count'
df = pd.read_csv('sipri-report-explosions.csv')

# Pivot the data to create a DataFrame suitable for a stacked bar chart
df_pivot = df.pivot_table(index='year', columns='country', values='Count', aggfunc='sum', fill_value=0)


# Plotting the data
plt.figure(figsize=(12, 8))
ax = df_pivot.plot(kind='bar', stacked=True)

# Set x-axis ticks every 5 years
ax.set_xticks(range(0, len(df_pivot.index), 5))
ax.set_xticklabels(df_pivot.index[::5])

plt.xlabel('Year')
plt.ylabel('Number of Nuclear Bombs Detonated')
plt.title('Nuclear Bombs Detonated per Year and Country')
plt.legend(title='Country', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```
